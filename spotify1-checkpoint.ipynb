```python
spotify['key'].value_counts()
```

```python
key_mapping = {0.0: 'C', 1.0: 'C♯,D♭', 2.0: 'D', 3.0: 'D♯,E♭', 4.0: 'E', 5.0: 'F', 6.0: 'F♯,G♭', 7.0: 'G', 8.0: 'G♯,A♭', 9.0: 'A', 10.0: 'A♯,B♭', 11.0: 'B'}

spotify['key'] = spotify['key'].map(key_mapping)

sns.countplot(x =spotify['key'],data=spotify, order=spotify['key'].value_counts().index)
plt.title("Count of Song Keys")
plt.show()

#So we can see most popular tracks are written in the key of C♯/D♭.

```

```python
#print mean value for mode
print("Mean value for mode feature:", spotify['mode'].mean())

#map the binary value of mode to major/minor
mode_mapping = {1.0: "major", 0.0: "minor"}
spotify['mode'] = spotify['mode'].map(mode_mapping)

#draw a countplot of the values
sns.countplot(x = 'mode', data=spotify)
plt.title("Count of Major/Minor Songs")
plt.show()

#Hence We can see 61 Percent of songs were in Major mode which creates a happpy sounding while Minor mode is assocaited with sad sound.
```

```python
#mean value and distplot for speechiness feature
print("Mean value for speechiness:", spotify['speechiness'].mean())
sns.distplot(spotify['speechiness'])
plt.show()

#The mean value for speechiness is pretty low (only 0.113). This only reiterates that people tend to listen to mostly music 
#on spotify compared to podcasts and audiobooks
```
